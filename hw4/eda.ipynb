```python
# Function to detect outliers using the IQR rule
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers

# Loop through numeric columns and report stats
for col in numeric_cols:
    print(f"\n==== {col} ====")
    print(f"Count     : {df[col].count()}")
    print(f"Min       : {df[col].min()}")
    print(f"Max       : {df[col].max()}")
    print(f"Mean      : {df[col].mean():.2f}")
    print(f"Median    : {df[col].median():.2f}")
    
    outliers = detect_outliers(df[col].dropna())
    print(f"Outliers  : {len(outliers)} found")
    
    if len(outliers) > 0:
        print(f"  - Lower bound: {outliers.min():.2f}")
        print(f"  - Upper bound: {outliers.max():.2f}")
    else:
        print("  - No significant outliers detected.")
```

```python
# Create crosstab
ct = pd.crosstab(df['Pclass'], df['Embarked'])

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(ct, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passenger Class vs. Embarkation Port")
plt.xlabel("Embarked")
plt.ylabel("Pclass")
plt.tight_layout()
plt.show()
```
